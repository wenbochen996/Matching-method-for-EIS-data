```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
# 读取CSV文件
data = pd.read_csv(r'D:\金属腐蚀\程序\选择模型\新数据集 23.08.04\Merged_file.csv')

# 获取阻抗数据
impedance_data = data.iloc[:, :-2].values

# 计算相关系数矩阵
corr_matrix = pd.DataFrame(impedance_data).corr()
fig = plt.subplots(figsize=(100, 100))
# 绘制热力图
plt.imshow(corr_matrix, cmap='coolwarm')
# 添加颜色条
plt.colorbar()
# 设置标题和标签
plt.title('Correlation Matrix')
plt.xlabel('Features')
plt.ylabel('Features')
```

```python
def xyname(start_num,end_num):
    index_ls = []
    scale_ls = []
    for i in range(start_num,end_num):
        index_ls.append(data.columns[i])
        scale_ls.append(i+1)
    return index_ls,scale_ls
```

```python
result1 = corr_matrix.iloc[:22,:22]
fig = plt.subplots(figsize=(10, 10))
index_ls,scale_ls = xyname(0,22)
# 绘制热力图
plt.imshow(result1, cmap='coolwarm')

# 添加颜色条
plt.colorbar()
plt.xticks(scale_ls,index_ls,rotation=90)
plt.yticks(scale_ls,index_ls)
# 设置标题和标签
plt.title('Correlation Matrix')
plt.xlabel('Features')
plt.ylabel('Features')

# 显示图形
plt.show()
```
